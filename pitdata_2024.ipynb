```python
!pip install pybaseball pandas
from datetime import datetime, timedelta
from pybaseball import statcast
import pandas as pd

years = range(2020, 2025)  # 2020〜2024
months = range(3, 12)      # 3月〜11月（10月・11月はプレーオフ）

for year in years:
    for month in months:
        start_date = datetime(year, month, 1)
        # 月末日計算（最終日は当月+1日の1日前 or 9/30, 11/30等）
        if month == 12:
            continue  # 12月はやらない
        if month == 11:
            last_day = datetime(year, month, 30)
        else:
            next_month = datetime(year, month + 1, 1)
            last_day = next_month - timedelta(days=1)

        # 現在の年の9月30日超えたら、その年の最後は11月30日まで
        if last_day > datetime(year, 11, 30):
            last_day = datetime(year, 11, 30)

        print(f"{start_date.date()} ~ {last_day.date()} ダウンロード開始")
        df = statcast(start_date.strftime('%Y-%m-%d'), last_day.strftime('%Y-%m-%d'))
        df.to_csv(f'statcast_{year}_{month:02}.csv', index=False)
        print(f"{year}-{month:02} 完了")
```

```python
import pandas as pd
import glob

files = sorted(glob.glob("statcast_*.csv"))
df_list = []

for file in files:
    try:
        df = pd.read_csv(file)
        if df.empty or df.columns.size == 0:
            print(f"{file} はデータなし or カラムなし。スキップ")
            continue
        df_list.append(df)
    except pd.errors.EmptyDataError:
        print(f"{file} は完全に空。スキップ")

df_all = pd.concat(df_list, ignore_index=True)
print("結合完了！")
```
